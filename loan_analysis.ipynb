```python
df = pd.read_csv('dataset_updated.csv')
cols = df.columns[6:]
d_cols = ['NCA-diff', 'CA-diff', 'TA-diff', 'CL-diff', 'TL-diff', 'SC-diff', 'RE-diff', 'Net-diff', 'Rev-diff', 'NPBT-diff', 'NPAT-diff']

for c in cols:
    df = df[df[c] != 'n.a.']
    df[c] = df[c].astype('float64')
    
df
```

```python
# df['NCA-diff'] = df['Non-Current Assets 2019'] - df['Non-Current Assets 2018']
# df['NCA-diff']
for i in range(0, len(cols), 2):
    df[d_cols[i//2]] = df[cols[i]] - df[cols[i+1]]
df
```

```python
l = df.columns.tolist()
l[2] = 'Loan Status'
l[3] = 'Paid-up Capital'
l

df = df.reindex(columns=l)
```

```python
df = df.drop(['Year of Incorporated', 'Industry'], axis=1)
```

```python
df = df.drop(cols, axis=1)
```

```python
X = df.iloc[:, 1:].to_numpy()
X.shape
```

```python
Y = df.iloc[:, 0].to_numpy()
Y = list(map(lambda x: 0 if x == 'BAD' else 1,Y))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, random_state=7, test_size=0.25)
```

```python
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import RobustScaler
import statsmodels.api as sm

scaler = preprocessing.RobustScaler() # Features Scaling is required for distance-based algorithms
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
import xgboost as xgb
from sklearn.ensemble import AdaBoostRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR

from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import GridSearchCV
```

```python
# Random Forests
regr_rf = RandomForestRegressor()
regr_rf.fit(X_train, y_train)
regr_rf.score(X_test, y_test)
```

```python
regr_rf.fit(X_train_scaled, y_train)
regr_rf.score(X_test_scaled, y_test)
```

```python
regr_rf.feature_importances_
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(random_state=7)
clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```

```python
clf.fit(X_train_scaled, y_train)
clf.score(X_test_scaled, y_test)
```

```python
from sklearn.metrics import plot_confusion_matrix
import matplotlib.pyplot as plt

print('Bad: {}, Good: {}'.format(Y.count(0), Y.count(1)))
plot_confusion_matrix(clf, X, Y, display_labels=['Bad', 'Good'], cmap=plt.cm.Blues)
```
